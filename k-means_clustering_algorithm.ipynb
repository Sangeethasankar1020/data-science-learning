### How K-Means Clustering Works 


K-Means is an unsupervised machine learning algorithm that groups data into K clusters based on feature similarity. Here's how it works step by step:



Step 1: Choose the Number of Clusters (K) You decide on a value of K (number of clusters). Choosing the right K is important (we'll discuss how to do this later). 

 Step 2: Initialize Cluster Centers (Centroids) The algorithm randomly picks K points from the dataset as the initial centroids. Each centroid represents the center of a cluster.

Step 3: Assign Data Points to the Nearest Centroid For each data point, calculate the distance to all centroids. Assign the point to the nearest centroid. This step forms K clusters. 

Step 4: Recalculate Centroids Compute the average of all points in each cluster. The new centroid is the mean of these points.

Step 5: Repeat Until Centroids Don’t Change Steps 3 and 4 are repeated until: Centroids stop changing significantly. The clusters become stable. This ensures the best grouping of data. Example: Clustering People by Age & Salary If we have a dataset with Age and Salary, K-Means will:

Group people into K different salary-age segments. Each group will have similar spending power or purchasing behavior. Businesses can target different customer segments based on clusters. How to Choose the Best K? To determine the optimal K, we use:

### Elbow Method 📐 Plot the Within-Cluster Sum of Squares (WCSS) vs. K values. The point where the WCSS graph bends like an elbow is the best K. Silhouette Score Measures how well each point is grouped. Higher score = better clustering

K-Means Pros & Cons 

✅ Advantages:

- Simple & fast. 
- Works well on large datasets. 
- Easy to interpret results.

⚠️ Disadvantages:

- Sensitive to outliers.
- Needs K to be predefined.
- Doesn’t work well with non-spherical clusters.

### Using Elbow method

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Step 1: Load the sample data
df = pd.read_excel("sample_data.xlsx")
```

```python
# Step 2: Selecting relevant features for clustering
X = df[['Age', 'Salary']]
```

```python
X
```

```python
# Step 3: Standardize the features to ensure proper clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 4: Finding the optimal number of clusters (K) using the Elbow Method
wcss = []  # Within-cluster sum of squares
for k in range(1, 11):  # Checking for K values from 1 to 10
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)  # Append the inertia (sum of squared distances)
```

```python
# Step 5: Plot the Elbow Method graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
# Step 6: Apply K-Means clustering with optimal K (assume K=3 based on the elbow method)
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Step 7: Plot the clustered data
plt.figure(figsize=(8, 6))
plt.scatter(X['Age'], X['Salary'], c=df['Cluster'], cmap='viridis', edgecolors='k')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('K-Means Clustering on Age vs Salary')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Step 8: Print the first few rows of clustered data
print(df.head())
```

# Model Evaluation 

- pandas: Handles data in tabular format.
- numpy: Used for numerical operations.
- matplotlib.pyplot: Plots graphs to visualize model performance.
- sklearn.model_selection.train_test_split: Splits the dataset into training and testing sets.
- sklearn.linear_model.LinearRegression: Implements Linear Regression for prediction.
- sklearn.metrics: Evaluates the model using error metrics like MAE, RMSE, and R¬≤ Score.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
data =pd.read_excel("house_prices.xlsx")
```

```python
df =pd.DataFrame(data)
```

```python
df
```

```python
# üéØ Features & Target
```

- X (Features): The independent variables used for predictions (Size, Bedrooms, Location Score).
- y (Target Variable): The dependent variable (Price) that we are trying to predict.

```python
X = df[['Size (sqft)', 'Bedrooms', 'Location Score']]
Y=df["Price ($1000s)"]
```

```python
# üèãÔ∏è Split Data into Train & Test
```

- train_test_split() splits the dataset into:
- 80% Training Data (X_train, y_train) ‚Üí Used to train the model.
- 20% Testing Data (X_test, y_test) ‚Üí Used to evaluate performance.
- test_size=0.2: Reserves 20% of the data for testing.
- random_state=42: Ensures reproducibility (same random split every time).

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# üèóÔ∏è Train the Model
```

- LinearRegression() creates a linear regression model.
- .fit(X_train, y_train) trains the model using the training dataset.

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#make prediction
```

```python
y_pred = model.predict(X_test)
```

- The trained model predicts house prices using X_test.
- y_pred stores the predicted prices.

### Evaluating Model Performance

```python
mae = mean_absolute_error(y_test, y_pred)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
r2 = r2_score(y_test, y_pred)
```

- Mean Absolute Error (MAE): Measures average prediction error. Lower = better.
- Root Mean Squared Error (RMSE): Penalizes large errors. Lower = better.
- R¬≤ Score: Measures how well the model explains variance in data.
- 1.0 ‚Üí Perfect Model
- 0.8 - 1.0 ‚Üí Good Model
- 0.5 - 0.8 ‚Üí Moderate Model
- Below 0.5 ‚Üí Poor Model

```python
print(f"üìå Mean Absolute Error (MAE): {mae:.2f}")
print(f"üìå Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"üìå R¬≤ Score: {r2:.2f}")
```

### Visualizing the model's performance

```python
plt.scatter(y_test, y_pred, color="blue")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle="--", color="red")
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted House Prices")
plt.show()
```

- Scatter Plot: Shows how close predictions (y_pred) are to actual values (y_test).
- Red Line: Represents a perfect prediction line (y = x).
- If points are close to the line, the model is accurate.

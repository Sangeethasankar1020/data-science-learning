# Feature Engineering 

### What is Feature Engineering?

Feature engineering is the process of:

- Selecting

- Transforming

- Creating new features from raw data

It helps ML models capture the underlying patterns in data more effectively.

**Why is Feature Engineering Important?**

- Raw data is not always in the best form for ML models.

- The right features can significantly improve model accuracy.

- Reduces overfitting by selecting relevant features.

- Improves interpretability of the model.

**Use Case:**

Imagine you are working on a house price prediction model. Raw data contains features like:

- Square footage

- Number of bedrooms

- Distance to the city center

However, you might need to engineer new features like:

- Price per square foot

- Age of the house

- Proximity to schools


# Transforming Predictors (Feature Transformation)

Feature transformation modifies existing features to make them more useful.

**Techniques:**

1.Scaling (Normalization & Standardization)

- Normalization scales features between 0 and 1.

- Standardization makes data have zero mean and unit variance.

2.Encoding Categorical Variables

- One-hot encoding: Converts categorical values into binary variables.

- Label encoding: Assigns a unique integer to each category.

3.Handling Missing Values

- Mean/Median/Mode imputation for numerical data.

- Filling with 'Unknown' for categorical data.

4.Log Transformation

- Used when a feature is highly skewed.

- Converts exponential data into a linear scale.

**Use Case:**
- Normalization ensures that numerical features are on the same scale.

- One-hot encoding converts categorical features into a machine-readable format.



# Feature Selection Methods

Feature selection helps identify the most important features for a model.

**Why Feature Selection?**
- Improves model efficiency.

- Reduces computation time.

- Prevents overfitting by removing redundant features.

**Types of Feature Selection Methods:**

1.Filter Methods

- Uses statistical tests like correlation or chi-square to select features.

2.Wrapper Methods

- Uses a model (e.g., Random Forest) to determine the best features.

3.Embedded Methods

- Feature importance is built into ML algorithms (like Lasso Regression).

**Use Case:**

- Selects only the most relevant features, reducing computation load.

- Helps in improving model accuracy.

# Backward Elimination Technique

Backward elimination is a stepwise regression technique that removes the least significant features.

**How it Works?**

- Start with all features.

- Fit a model and check feature importance.

- Remove the least significant feature.

- Repeat until all remaining features are statistically significant.

**Use Case:**
- Eliminates unnecessary features, making the model simpler.

- Helps understand which features are statistically significant.

# Feature Importance from ML Modeling

Feature importance helps us understand which features contribute most to the model.

**How is Feature Importance Measured?**

Tree-based models (Random Forest, XGBoost) have built-in feature importance.

Coefficients in regression models indicate feature importance.



**Use Case:**

Helps in feature selection by identifying the most important features.

Improves model performance by focusing on relevant data.
